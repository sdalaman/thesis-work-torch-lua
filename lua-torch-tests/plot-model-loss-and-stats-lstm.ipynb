```python
def plotLossLstm(fname):
    fin = open(fname, 'r')
    data = fin.read()
    lines = data.split('\n')
    epoch = []
    lr = []
    loss = []
    lossPriMean = []
    lossSecMean = []
    lossMean = []
    head = lines[0].split(",")[0]
    for line in lines[2:len(lines)]:
        flds = line.split(",")
        if len(flds) == 6:
            epoch.append(flds[0])
            loss.append(flds[2])
            lossPriMean.append(flds[3])
            lossSecMean.append(flds[4])
            lossMean.append(flds[5])
    print(len(epoch),len(loss),len(lossPriMean),len(lossSecMean),len(lossMean))
    plt.gca().set_prop_cycle=(cycler('color', ['r', 'g', 'b', 'y']))
    plt.plot(np.array(epoch), np.array(loss))
    plt.plot(np.array(epoch), np.array(lossPriMean))
    plt.plot(np.array(epoch), np.array(lossSecMean))
    plt.plot(np.array(epoch), np.array(lossMean))
    plt.legend(['err','priMean','secMean','mean'], loc='upper right')
    plt.title(head)
    plt.xlabel('epoch')
    plt.ylabel('error')
    plt.show()
```

```python
def plotLossTanh(fname):
    fin = open(fname, 'r')
    data = fin.read()
    lines = data.split('\n')
    epoch = []
    mean = []
    meanT = 0
    cnt = 0
    loss = []
    head = lines[0].split(",")[1]
    for line in lines:
        flds = line.split(",")
        if len(flds) == 3:
            epoch.append(flds[0])
            loss.append(flds[2])
            meanT += float(flds[2])
            cnt += 1
            mean.append(meanT/cnt)
    print(len(epoch),len(loss))
    plt.gca().set_prop_cycle=(cycler('color', ['r', 'g', 'b', 'y']))
    plt.plot(np.array(epoch), np.array(loss))
    plt.plot(np.array(epoch), np.array(mean))
    plt.legend(['err','mean'], loc='upper right')
    plt.title(head)
    plt.xlabel('epoch')
    plt.ylabel('error')
    plt.show()
```

```python
plotLossTanh('test_errors.new2.tanh.en-tr-0.1.txt')
```

```python
def plotLossAdd(fname):
    fin = open(fname, 'r')
    data = fin.read()
    lines = data.split('\n')
    epoch = []
    mean = []
    meanT = 0
    cnt = 0
    loss = []
    head = ""
    for line in lines:
        flds = line.split(",")
        if len(flds) == 2:
            epoch.append(flds[0])
            loss.append(flds[1])
            meanT += float(flds[1])
            cnt += 1
            mean.append(meanT/cnt)
    print(len(epoch),len(loss))
    plt.gca().set_prop_cycle=(cycler('color', ['r', 'g', 'b', 'y']))
    plt.plot(np.array(epoch), np.array(loss))
    plt.plot(np.array(epoch), np.array(mean))
    plt.legend(['err','mean'], loc='upper right')
    plt.title(head)
    plt.xlabel('epoch')
    plt.ylabel('error')
    plt.show()
```
